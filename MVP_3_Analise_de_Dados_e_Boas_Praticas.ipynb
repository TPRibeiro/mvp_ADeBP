# Introdução

A Organização Mundial da Saúde (OMS) define obesidade como o acúmulo excessivo de gordura corporal que pode prejudicar a saúde (WHO, 2024).

A obesidade é considerada uma doença crônica que afeta pessoas de todas as idades e grupos sociais. Ela pode levar a doenças crônicas como diabetes tipo 2, doenças cardiovasculares, hipertensão, acidente vascular cerebral e vários tipos de câncer.

Indivíduos em risco são orientados por profissionais de saúde a comer menos calorias e se exercitar mais, frequentemente usando limites de índice de massa corporal (IMC; peso/altura²) para triagem e para orientar o progresso e o prognóstico. A OMS considera uma pessoa obesa quando seu Índice de Massa Corporal (IMC) é igual ou superior a 30 kg/m².

De acordo com a Organização Pan-americana de Saúde (PAHO), uma em cada oito pessoas, no mundo, vive com obesidade (2024), o que demonstra a importância da prevenção e do controle da obesidade desde o início da vida até a idade adulta.

A obesidade é uma doença crônica complexa, contudo, as causas são bem compreendidas, assim como as intervenções necessárias para conter a crise, que são respaldadas por fortes evidências. No entanto, elas não são implementadas (PAHO, 2024).

Este projeto tem como objetivo aplicar técnicas de ciência de dados na análise de um conjunto de dados que reúne informações sobre indivíduos com diferentes perfis físicos e comportamentais.

Embora o objetivo final seja a construção de um modelo preditivo capaz de estimar a categoria de obesidade de uma pessoa com base em variáveis mais abrangentes do que apenas altura e peso (como o IMC), este trabalho não abrangerá a etapa de modelagem. O foco estará na compreensão de como fatores como alimentação, prática de atividade física, consumo de água e hábitos de vida se correlacionam com os diferentes níveis de obesidade.

A análise e o pré-processamento desses dados permitirão, nas próximas etapas, a construção de visualizações, extração de padrões e preparação para a modelagem preditiva. Ao longo do projeto, serão aplicadas boas práticas de análise de dados com foco em limpeza, transformação, validação de hipóteses e justificativas técnicas para cada escolha tomada.

## Hipóteses do Problema

Este projeto considera múltiplos fatores comportamentais, físicos e demográficos que podem influenciar o risco de desenvolvimento de obesidade.

No dataset, estão presentes as 17 variáveis e 2111 instâncias. Abaixo estão listadas hipóteses específicas associadas a algumas variáveis-chaves presentes no dataset.


- **Hipótese 1:**
 *Indivíduos que monitoram a ingestão calórica com regularidade tendem a apresentar menor prevalência de obesidade, por estarem mais conscientes de seus hábitos alimentares.*

A variável `SCC` representa se a pessoa monitora ou não a quantidade de calorias consumidas. Espera-se que esse comportamento esteja associado a uma maior disciplina alimentar e, consequentemente, a menores níveis de obesidade.

- **Hipótese 2:**
 *Maior frequência de atividade física está associada à menor prevalência de obesidade.*

A prática de atividade física é um dos principais fatores de proteção contra a obesidade. A variável `FAF` representa a quantidade de horas de atividade física praticada semanalmente.

- **Hipótese 3:**
 *Pessoas que consomem mais água diariamente têm menor risco de desenvolver obesidade.*

A hidratação adequada está associada a processos metabólicos mais eficientes e pode influenciar positivamente no controle do apetite e da ingestão calórica. A variável `CH2O` indica a quantidade de água consumida por dia (em litros).

- **Hipótese 4:**
 *Indivíduos que consomem fast food com maior frequência têm maior propensão à obesidade.*

A variável `FCVC` representa a frequência com que os indivíduos consomem alimentos calóricos, como fast food. A associação entre consumo excessivo de fast food e aumento do risco de obesidade é amplamente documentada na literatura científica.

- **Hipótese 5:**
 *Indivíduos com histórico familiar de sobrepeso têm maior predisposição a desenvolver obesidade, devido a fatores genéticos e comportamentais compartilhados.*

A variável `family_history_with_overweight` indica se o indivíduo possui familiares com histórico de excesso de peso. Esse fator é importante do ponto de vista epidemiológico e preventivo, pois pode sinalizar uma predisposição biológica ou cultural à obesidade.






## Tipo do problema

Tendo em vista o objetivo da construção de um modelo preditivo capaz de estimar a categoria de obesidade, este trata-se de um problema de aprendizado supervisionado, uma vez que o conjunto de dados utilizado contém uma variável-alvo conhecida (NObeyesdad), que representa a classificação do nível de obesidade de cada indivíduo. ##Atributos do dataset

O dataset contém os seguintes atributos:

* Gender - Gênero do indivíduo (Male ou Female)
* Age	- Idade do indivíduo
* Height - Altura (em metros)
* Weight -	Peso (em kg)
* family_history_with_overweight - Histórico familiar de sobrepeso (yes/no)
* FAVC - Consumo frequente de alimentos com alto valor calórico (yes/no)
* FCVC -	Frequência de consumo de vegetais (escala de 1 a 3)
* NCP -	Número de refeições principais por dia
* CAEC -	Consumo de alimentos entre as refeições (no, Sometimes, Frequently, Always)
* SMOKE -	Fuma (yes/no)
* CH2O -	Consumo de água por dia (litros)
* SCC	- Monitora consumo de calorias (yes/no)
* FAF -	Frequência de atividade física por semana (horas)
* TUE -	Tempo usando dispositivos tecnológicos por dia (horas)
* CALC -	Frequência de consumo de álcool (no, Sometimes, Frequently, Always)
* MTRANS -	Meio de transporte utilizado com mais frequência
* NObeyesdad -	Variável alvo (target): nível de obesidade (ex: Normal Weight, Overweight Level I, Obesity Type I, etc.)

Adicionalmente, os valores atribuídos aos níveis de obesidade se basearam no Índice de Massa Corporal (IMC) do usuário e na respectiva classificação utilizada pela OMS, que se dá da seguinte forma:

* IMC < 18,5 kg/m² - baixo peso
* IMC > 18,5 até 24,9 kg/m² - peso adequado
* IMC ≥ 25 até 29,9 kg/m² - sobrepeso
* IMC > 30,0 kg/m² até 34,9 kg/m² - obesidade grau 1
* IMC > 35 kg/m² até 39,9 kg/m² - obesidade grau 2
* IMC > 40 kg/m² - obesidade extrema

## Estatísticas Descritivas dos Atributos Numéricos

Esta etapa tem como objetivo entender a distribuição dos dados numéricos presentes no dataset. Serão calculadas medidas de tendência central (média, mediana, moda), dispersão (desvio-padrão), além de valores mínimos, máximos e a quantidade de valores ausentes.

Essas estatísticas ajudam a identificar possíveis outliers, distorções ou padrões interessantes que podem influenciar etapas futuras como o pré-processamento ou a modelagem.

> Add blockquote



```python
# Selecionando apenas as colunas com dados numéricos (int ou float)
colunas_numericas = obesidade.select_dtypes(include=['int64', 'float64']).columns

# Criando um DataFrame com as estatísticas descritivas
estatisticas = pd.DataFrame(index=colunas_numericas)

# Calculando as principais métricas para cada atributo numérico
estatisticas['Mínimo'] = obesidade[colunas_numericas].min()
estatisticas['Máximo'] = obesidade[colunas_numericas].max()
estatisticas['Média'] = obesidade[colunas_numericas].mean()
estatisticas['Mediana'] = obesidade[colunas_numericas].median()
estatisticas['Moda'] = obesidade[colunas_numericas].mode().iloc[0]  # iloc[0] pega a primeira moda, caso haja múltiplas
estatisticas['Desvio-Padrão'] = obesidade[colunas_numericas].std()

# Exibindo o resultado em formato de tabela
display(estatisticas)
```

Dentre os valores encontrados, todos se encontram conforme o esperado, com a exceção do peso ("weight"), que possui um alto desvio-padrão e alguns valores mínimo e máximo possivelmente discrepantes quando se comparado com a média.

Assim, uma análise mais detalhada do peso será realizada para uma melhor avaliação. Realizaremos esta análise ao verificarmos o peso em um gráfico de dispersão ("scatter plot") e boxplot, que permitirá uma verificação facilitada dos valores discrepantes ("outliers").

```python
# Cálculo da média de peso
media_peso = obesidade['Weight'].mean()

# Criação do gráfico de dispersão
plt.figure(figsize=(12, 5))
plt.scatter(
    x=range(len(obesidade['Weight'])),       # Índices no eixo X
    y=obesidade['Weight'],                   # Pesos no eixo Y
    alpha=0.6,                               # Transparência
    edgecolors='black'                       # Contorno dos pontos
)

# Adição de uma linha horizontal na média
plt.axhline(y=media_peso, color='red', linestyle='--', linewidth=2,
            label=f'Média de Peso: {media_peso:.2f} kg')

# Rótulos e título
plt.title('Distribuição dos valores de peso (Weight)')
plt.xlabel('Índice do Registro')
plt.ylabel('Peso (kg)')
plt.legend()
plt.show()
```

```python
# Criação do boxplot para a variável Weight
plt.figure(figsize=(6, 5))
plt.boxplot(obesidade['Weight'], vert=True, patch_artist=True)

# Adiciona título e rótulo
plt.title('Boxplot dos Valores de Peso')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.show()
```

```python
# Cálculo dos quartis
q1 = obesidade['Weight'].quantile(0.25)  # 1º quartil
q3 = obesidade['Weight'].quantile(0.75)  # 3º quartil
iqr = q3 - q1                            # Intervalo interquartil

# Limites inferior e superior para outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

print(f"Q1 (25%): {q1:.2f} kg")
print(f"Q3 (75%): {q3:.2f} kg")
print(f"IQR: {iqr:.2f}")
print(f"Limite inferior: {limite_inferior:.2f} kg")
print(f"Limite superior: {limite_superior:.2f} kg")

# Filtrando os valores considerados outliers
outliers = obesidade[(obesidade['Weight'] < limite_inferior) |
                     (obesidade['Weight'] > limite_superior)]

# Exibindo os registros considerados outliers
print(f"\nTotal de outliers encontrados: {len(outliers)}")
display(outliers[['Age', 'Gender', 'Height', 'Weight']])
```

Durante a análise inicial dos valores de `Weight`, foi identificado um valor considerado outlier pelo boxplot: um indivíduo com 173 kg. Embora esse valor se destaque estatisticamente, ele pertence a uma pessoa de 18 anos com 1,83 m de altura — o que, apesar de indicar um quadro de obesidade severa, ainda se encontra dentro de limites fisiológicos plausíveis.

Por outro lado, o dataset também apresenta indivíduos com peso extremamente baixo, como 39 kg. Embora esse valor não tenha sido identificado como outlier pelo boxplot univariado de `Weight`, ele levanta sérias dúvidas quando se considera o restante do contexto, como idade e altura. De acordo com os dados, a idade média dos indivíduos no conjunto é de aproximadamente 24 anos e a altura média é de 1.70 m e, para essa faixa, um peso de 39 kg tende a ser incompatível com o de uma pessoa adulta saudável.

Por esse motivo, optou-se por realizar uma análise mais contextualizada, combinando as variáveis `Weight`, `Age` e `Height` em um gráfico de dispersão multivariado. O objetivo é observar a distribuição do peso em relação à idade, com a altura sendo representada por uma escala de cores. Essa visualização mais rica possibilita identificar padrões incomuns e detectar valores realmente discrepantes quando considerados múltiplos fatores fisiológicos ao mesmo tempo.


```python
# Cria o gráfico de dispersão
plt.figure(figsize=(12, 6))

# Cada ponto representa uma pessoa
scatter = plt.scatter(
    x=obesidade['Age'],               # Idade no eixo X
    y=obesidade['Weight'],            # Peso no eixo Y
    c=obesidade['Height'],            # Altura representada pela cor
    cmap='viridis',                   # Paleta de cores agradável
    edgecolors='black',
    alpha=0.8
)

# Adiciona barra de cor (legenda da altura)
cbar = plt.colorbar(scatter)
cbar.set_label('Altura (m)')

# Rótulos e título
plt.title('Peso vs. Idade com Altura como Cód. de Cor')
plt.xlabel('Idade (anos)')
plt.ylabel('Peso (kg)')

# Grade para facilitar leitura visual
plt.grid(True)
plt.show()
```

Com base na análise multivariada por meio do gráfico de dispersão entre `Weight`, `Age` e `Height`, foi possível identificar padrões mais realistas e coerentes dentro do conjunto de dados.

Indivíduos com peso muito elevado (acima de 160 kg) foram contextualizados como pessoas jovens com altura acima da média, o que embora indique obesidade grave, ainda pode ser biologicamente plausível.

Similarmente, valores muito baixos de peso (como 39 kg) estão associados com alturas mais baixas e, apesar de indicarem pesos abaixo do saudável, ainda são valores biologicamente plausíveis.

Ainda assim, optou-se por realizar uma nova análise para determinarmos exatamente os valores relacionados. Dessa forma, uma análise para melhor identificar a presença de valores biologicamente inválidos será realizada no próximo passo.

```python
# Define limites considerados biologicamente plausíveis
peso_min = 30         # valores abaixo de 30 kg são considerados perigosamente baixos para adultos
peso_max = 160        # valores acima de 160 kg são extremos, mesmo em casos de obesidade severa
altura_min = 1.35     # alturas abaixo de 1.35m são atípicas para adultos
altura_max = 2.10     # alturas acima de 2.10m são muito raras
idade_min = 15        # idades abaixo de 15 anos indicam população infantojuvenil
idade_max = 100       # por segurança, foram excluídos valores potencialmente anômalos acima de 100

# Cria uma visualização dos registros que violam pelo menos um dos critérios acima
valores_invalidos = obesidade[
    (obesidade['Weight'] < peso_min) |
    (obesidade['Weight'] > peso_max) |
    (obesidade['Height'] < altura_min) |
    (obesidade['Height'] > altura_max) |
    (obesidade['Age'] < idade_min) |
    (obesidade['Age'] > idade_max)
]

# Exibe os valores considerados possivelmente inválidos
valores_invalidos
```

Os critérios utilizados para essa análise foram:

- **Peso (Weight):** valores abaixo de 30 kg ou acima de 160 kg foram considerados extremos.
- **Altura (Height):** valores abaixo de 1,35 m ou acima de 2,10 m foram considerados biologicamente raros.
- **Idade (Age):** registros com idade inferior a 15 anos ou superior a 100 anos foram sinalizados como atípicos.

Essa análise resultou na identificação de 5 registros possivelmente inválidos. No entanto, foi observado que, apesar de inicialmente parecerem outliers extremos, os registros apresentam combinações plausíveis de altura, idade e peso.

Dessa forma, nenhum valor será removido na etapa de pré-processamento de dados por motivo de discrepância. A decisão visa preservar a integridade e a representatividade do conjunto de dados, especialmente considerando que muitos dos dados são sintéticos.

## Análise de Correlação entre Variáveis Contínuas

Nesta etapa, será gerado um *heatmap* (mapa de calor) com o objetivo de identificar possíveis correlações entre as variáveis contínuas do dataset. A análise de correlação é uma técnica estatística fundamental que mede o grau de associação entre duas variáveis numéricas. Seu principal uso é revelar relações lineares que podem indicar redundância, dependência ou padrões relevantes nos dados.

No contexto deste projeto, o *heatmap* será construído com base nas seguintes variáveis normalizadas: `Age`, `Height`, `Weight`, `FCVC`, `NCP`, `CH2O`, `FAF` e `TUE`. Estas variáveis representam medições contínuas relacionadas a idade, hábitos alimentares e estilo de vida dos indivíduos.

A escolha pelo uso do *heatmap* se justifica por diversos fatores:

- **Visualização clara e intuitiva** de correlações positivas ou negativas entre variáveis, através de uma escala de cores.
- **Detecção de relações relevantes** entre os atributos que possam influenciar na obesidade, como, por exemplo, a associação entre frequência de atividade física e peso.
- **Identificação de possíveis colinearidades** que devem ser observadas futuramente durante a modelagem, especialmente em algoritmos sensíveis a esse fenômeno.

Essa análise complementa a exploração dos dados ao oferecer uma visão consolidada das interações entre atributos numéricos, o que pode também guiar escolhas futuras de engenharia de variáveis e seleção de características.

```python
# Seleciona apenas as variáveis numéricas contínuas do dataset
variaveis_continuas = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

# Calcula a matriz de correlação entre essas variáveis
correlacao = obesidade[variaveis_continuas].corr()

# Cria um gráfico de calor (heatmap) com anotações numéricas
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Título e layout
plt.title('Mapa de Correlação entre Variáveis Contínuas', fontsize=14)
plt.tight_layout()
plt.show()
```

### Interpretação do Heatmap de Correlação

O heatmap acima apresenta o coeficiente de correlação de Pearson entre as variáveis contínuas do dataset. Esse coeficiente varia de -1 a +1, onde:

- `+1` indica correlação perfeitamente positiva (quando uma variável aumenta, a outra também aumenta na mesma proporção),
- `-1` indica correlação perfeitamente negativa (quando uma variável aumenta, a outra diminui),
- `0` indica ausência de correlação linear.

Cada célula do gráfico mostra a correlação entre um par de variáveis. A cor azul representa correlações negativas, a cor vermelha representa correlações positivas, e a intensidade da cor reflete a força da correlação.

 1. **Correlação Altamente Positiva**:

2. **Correlações Levemente Positivas**:

  - Isso sugere que quanto mais vegetais a pessoa consome, levemente maior tende a ser o peso. Esse resultado parece contraintuitivo, mas pode estar ligado ao fato de que pessoas com obesidade (inclusive após diagnóstico) são incentivadas a consumir mais vegetais como medida corretiva.
  
- `Height` e `FAF`: **0.29**
  - Pessoas mais altas tendem a relatar mais atividade física semanal, embora a correlação ainda seja fraca. Pode haver um viés comportamental ou de geração sintética nesse padrão.

- `CH2O` (água) tem correlação fraca positiva com quase tudo, principalmente com `Height` (**0.21**) e `Weight` (**0.20**). Novamente, pode refletir tendências de consumo mais saudável adotadas por pessoas com maior estatura ou em contextos de saúde monitorada.

### Etapa 2 – Codificação de Variáveis

#### Etapa 2A - Codificação de Variáveis Categóricas

Após a verificação de valores ausentes, será realizada a codificação das variáveis categóricas. Esse processo é essencial porque algoritmos de aprendizado de máquina não conseguem trabalhar diretamente com variáveis em formato textual (`strings`).

Nesta etapa, será adotada a estratégia de One-Hot Encoding (OHE) para as variáveis nominais (sem ordem implícita):
- `Gender` — gênero
- `family_history_with_overweight` — histórico familiar de sobrepeso
- `FAVC` — consome alimentos altamente calóricos com frequência?
- `SMOKE` — fuma?
- `SCC` — monitora calorias consumidas?
- `CALC` — frequência de consumo de álcool
- `MTRANS` — meio de transporte utilizado
- `CAEC` — frequência de consumo entre refeições

Essas variáveis serão convertidas em colunas binárias (0 ou 1) para evitar que o modelo atribua significado ordinal a categorias que não o possuem.

```python
# Etapa 2a - Codificação das Variáveis Categóricas

# Lista das variáveis categóricas NOMINAIS (sem ordem entre os valores)
variaveis_nominais = [
    'Gender',                          # Gênero biológico
    'family_history_with_overweight', # Histórico familiar de sobrepeso
    'FAVC',                            # Consome alimentos altamente calóricos com frequência?
    'SMOKE',                           # Fuma?
    'SCC',                             # Monitora consumo de calorias?
    'CALC',                            # Frequência de consumo de álcool
    'MTRANS',                          # Meio de transporte utilizado
    'CAEC'                             # Frequência de consumo entre refeições
]

# Aplicação do One-Hot Encoding nas variáveis nominais, transformando cada valor distinto em uma nova coluna binária (0 ou 1).
# O argumento drop_first=True remove a primeira categoria de cada variável categórica. Isso evita que todas as colunas binárias (dummies) somem 1, o que causaria multicolinearidade — ou seja,
# uma relação linear perfeita entre as variáveis, o que prejudica modelos como Regressão Linear ou Logística.
# Ao remover uma coluna, mantemos a informação completa, mas eliminamos a redundância estrutural.
obesidade_codificada = pd.get_dummies(obesidade, columns=variaveis_nominais, drop_first=True)

# Exibe as primeiras linhas do DataFrame codificado para inspeção
obesidade_codificada.head()
```

### Etapa 3 – Normalização das Variáveis Numéricas

Os algoritmos de machine learning podem ser sensíveis à escala dos dados, especialmente aqueles que utilizam distâncias, gradientes ou otimização, como K-Nearest Neighbors (KNN), Regressão Logística, Redes Neurais e Support Vector Machines (SVM). Quando variáveis possuem magnitudes diferentes — como neste dataset, em que a altura varia entre 1.45 e 2.00 metros, e o peso pode ultrapassar 150 kg — os modelos podem interpretar erroneamente que os atributos de maior escala são mais relevantes apenas por seus valores absolutos.

Para evitar esse viés e garantir um melhor desempenho dos modelos, será aplicada a normalização Min-Max, que transforma os valores numéricos para uma escala entre 0 e 1, mantendo a distribuição original dos dados. Esta abordagem é simples, eficiente e facilita a comparação entre variáveis com unidades diferentes.

A escolha do Min-Max Scaling se deu por diversos fatores:

- **Interpretação direta**: os valores normalizados permanecem na mesma ordem e proporção relativa, o que facilita a interpretação e evita distorções nos dados.
- **Compatibilidade com dados sintéticos**: como parte do dataset foi gerada artificialmente por técnicas de oversampling (SMOTE), o Min-Max Scaling se mostra mais adequado, pois mantém a coerência dos valores contínuos sem alterar sua distribuição.
- **Consistência entre variáveis**: o uso do Min-Max em todas as variáveis contínuas garante que estejam na mesma escala, algo fundamental para algoritmos que utilizam medidas de distância ou otimização por gradiente.
- **Maior robustez prática**: diferente do Z-score (padronização), o Min-Max não assume nenhuma distribuição específica dos dados, sendo mais apropriado quando as variáveis não seguem uma distribuição normal — como é o caso neste projeto.
- **Evita distorções desnecessárias**: técnicas alternativas como o Robust Scaling, que se baseia na mediana e no intervalo interquartil, são indicadas quando há muitos outliers. No presente caso, após análise estatística e visual (boxplots e dispersões), verificou-se que os dados estão dentro de uma faixa coerente com a realidade, não justificando abordagens mais robustas.

Assim, o Min-Max Scaling foi adotado como a técnica de normalização mais alinhada com as características do conjunto de dados, os objetivos do projeto e os algoritmos que poderão ser empregados na modelagem preditiva.

Neste projeto, as variáveis numéricas normalizadas serão: `Age`, `Height`, `Weight`, `FCVC`, `NCP`, `CH2O`, `FAF` e `TUE`. Todas serão transformadas utilizando a técnica de Min-Max Scaling. Essas variáveis são do tipo numéricas contínuas, representam medições do mundo real (como idade, altura e frequência de atividades) e podem assumir uma ampla gama de valores, incluindo valores decimais e com escalas diferentes entre si.

As demais variáveis do dataset são categóricas e já foram tratadas com One-Hot Encoding, resultando em colunas binárias com valores 0 ou 1. Como essas colunas já estão naturalmente na mesma escala, não há necessidade (e nem seria adequado) aplicar a normalização novamente. Além disso, a variável-alvo `Target`, por se tratar de um problema de classificação, não deve ser normalizada.

```python
# Lista de colunas numéricas que serão normalizadas
variaveis_numericas = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

# Cria o objeto MinMaxScaler
escalonador = MinMaxScaler()

# Aplica a transformação Min-Max e substitui os valores no dataframe
obesidade_preprocessada[variaveis_numericas] = escalonador.fit_transform(obesidade_preprocessada[variaveis_numericas])

# Exibe amostra para verificar resultado
obesidade_preprocessada[variaveis_numericas].head()
```

```python
# 1. Verificação de duplicatas
# Checa se há linhas duplicadas no dataset processado
duplicatas = obesidade_preprocessada.duplicated().sum()
print(f"Número de linhas duplicadas: {duplicatas}")

# 2. Verificação dos tipos de dados (dtype)
# Garante que cada coluna esteja com o tipo de dado esperado (float, int, etc.)
print("\nTipos de dados por coluna:")
print(obesidade_preprocessada.dtypes)

# 3. Verificação de variáveis normalizadas
# Confirma que as colunas numéricas normalizadas estão dentro do intervalo [0, 1]
# Define as colunas que foram normalizadas com MinMax
colunas_normalizadas = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

# Verifica se há valores fora da faixa esperada
for coluna in colunas_normalizadas:
    minimo = obesidade_preprocessada[coluna].min()
    maximo = obesidade_preprocessada[coluna].max()
    print(f"{coluna} — mínimo: {minimo:.4f}, máximo: {maximo:.4f}")

# 4. Visualização de uma amostra do DataFrame final
# Exibe as primeiras 5 linhas do dataset pré-processado para inspeção manual
print("\nAmostra do dataset pré-processado:")
display(obesidade_preprocessada.head())
```

Os resultados da validação indicam que:

- Todos os tipos de dados estão apropriados:

- Os intervalos das variáveis numéricas estão normalizados:
Foi verificado que todos os valores das variáveis contínuas variam entre 0 e 1, o que confirma que a técnica de Min-Max Scaling foi aplicada corretamente.

No entanto, identificou-se a presença de 24 linhas duplicadas, ou seja, registros que possuem exatamente os mesmos valores em todas as colunas. Esse tipo de duplicação pode afetar a distribuição das classes e comprometer a representatividade dos dados, especialmente considerando que o dataset foi parcialmente gerado por técnicas de oversampling (SMOTE).

Com isso, embora a estrutura dos dados esteja tecnicamente válida e pronta para uso, a existência dessas duplicatas exige uma análise mais criteriosa para determinar se e como elas devem ser tratadas.

A próxima etapa, portanto, será dedicada à análise e tratamento dessas duplicações, com o objetivo de manter a qualidade e a diversidade informacional do dataset.



### 4A - Análise das duplicatas

Tendo em vista o resultado encontrado, a presença de 24 linhas duplicadas no dataset, uma análise mais detalhada é necessária para podermos determinar se estas podem ser mantidas ou devem ser removidas.

```python
# Filtra apenas as linhas duplicadas, mantendo a primeira ocorrência fora da lista
linhas_duplicadas = obesidade_preprocessada[obesidade_preprocessada.duplicated(keep=False)]

# Exibe todas as colunas das linhas duplicadas
print(f"Número total de duplicatas (incluindo a original): {linhas_duplicadas.shape[0]}")
display(linhas_duplicadas.sort_values(by=list(obesidade_preprocessada.columns)).reset_index(drop=True))
```

Durante a inspeção mais detalhadas dos dados duplicados, foram identificadas várias linhas completamente duplicadas no dataset - algumas repetidas mais de 10 vezes. Embora o uso da técnica SMOTE para geração sintética de dados possa justificar certa repetição, uma alta proporção de registros idênticos pode enviesar o modelo de aprendizado, levando a overfitting ou perda de generalização.

Para lidar com esse problema, será realizado um controle seletivo: será mantido um número máximo de duas ocorrências para cada grupo de duplicatas idênticas. Isso permite preservar parte da variação estatística introduzida pelo oversampling, sem que os dados se tornem artificialmente dominados por repetições.

O processo consiste em:
- Identificar todos os registros completamente idênticos (duplicatas exatas).
- Agrupar esses registros e limitar sua ocorrência a um máximo predefinido (`MAX_DUPLICATES = 2`).
- Manter todas as linhas únicas que ocorrem apenas uma vez.
- Concatenar os dados resultantes em um novo DataFrame final (`obesidade_limpa`), com uma estrutura balanceada e mais realista.

Esse procedimento garante que a base de dados esteja mais próxima de uma versão ideal para modelagem, reduzindo ruídos e vieses sem prejudicar a distribuição geral das classes e variáveis.

```python
# Define a quantidade máxima de vezes que queremos manter uma duplicata
MAX_DUPLICATES = 2

# Cria um DataFrame contendo todas as duplicatas (inclusive a primeira ocorrência de cada grupo duplicado)
duplicatas = obesidade_preprocessada[obesidade_preprocessada.duplicated(keep=False)]

# Agrupa pelas colunas completas para encontrar duplicatas exatas e mantém no máximo 'MAX_DUPLICATES' ocorrências por grupo
duplicatas_reduzidas = (
    duplicatas
    .groupby(list(obesidade_preprocessada.columns))
    .head(MAX_DUPLICATES)  # Mantém até o limite definido de cópias por grupo
)

# Obtém as linhas que ocorrem apenas uma vez no dataset (não estão duplicadas)
nao_duplicadas = obesidade_preprocessada.drop_duplicates(keep=False)

# Junta os dados únicos com as duplicatas reduzidas para compor o novo DataFrame final
obesidade_limpa = pd.concat([nao_duplicadas, duplicatas_reduzidas], ignore_index=True)

# Verifica quantas duplicatas ainda existem (idealmente nenhuma além do controle definido)
print("Número de duplicatas após redução:", obesidade_limpa.duplicated().sum())
print("Total de linhas após tratamento:", obesidade_limpa.shape[0])
```

O dataset utilizado foi descrito no artigo científico *“Dataset for estimation of obesity levels based on eating habits and physical condition in individuals from Colombia, Peru and Mexico”* (PALECHOR & MANOTAS, 2019), e contou com uma característica que percebeu-se relevante no desenvolvimento do projeto: aproximadamente 77% das instâncias foram geradas artificialmente por meio da técnica de oversampling SMOTE. - **Redução das duplicatas idênticas**, limitando sua ocorrência a no máximo duas por grupo, para evitar viés excessivo decorrente do oversampling.

Adicionalmente, confirmou-se a ausência de valores nulos, a correção dos tipos de dados, a adequação da escala das variáveis numéricas e a correta codificação das variáveis categóricas.

Conclui-se, portanto, que o dataset está devidamente preparado para ser utilizado em modelos preditivos, oferecendo uma base sólida para análises futuras.
